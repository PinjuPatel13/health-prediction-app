```python
import pandas as pd

# Step 1: Load your dataset
train_data = pd.read_csv('D:/User/Pinjal/Work/AI_Project/Medical_report_OCR/Data/trainn.csv')

# Check the first few rows of the data to make sure it's loaded properly
data = print(train_data.head())

# Check the column names and data types
# print(train_data.info())
data
```

```python
# Step 2: Split the data into features and labels
X_train = train_data.drop(columns=['prognosis'])  # Drop the target column
y_train = train_data['prognosis']  # The target column

# Check the shapes of the features and labels
print(X_train.shape)
print(y_train.shape)

```

```python
# Features (symptoms) and target (prognosis)
X_train = train_data.drop(columns=['prognosis'])  # Drop the target column
y_train = train_data['prognosis']  # The target column
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of each symptom
symptoms_count = X_train.sum(axis=0)
symptoms_count.plot(kind='bar', figsize=(12, 6))
plt.title("Distribution of Symptoms (Presence/Absence)")
plt.xlabel("Symptoms")
plt.ylabel("Count (Presence)")
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot the distribution of the target variable (health conditions)
y_train.value_counts().plot(kind='bar', figsize=(10, 6), color='orange')
plt.title("Distribution of Health Conditions (Prognosis)")
plt.xlabel("Health Conditions")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate the class distribution of the target variable
class_distribution = y_train.value_counts(normalize=True)
print("Class Distribution (Normalized):")
print(class_distribution)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import joblib

# Split data into training and testing sets (80% training, 20% testing)
X_train_split, X_test_split, y_train_split, y_test_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Initialize the model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train_split, y_train_split)

# Make predictions on the test set
y_pred = model.predict(X_test_split)

# Evaluate the model
print(classification_report(y_test_split, y_pred))

# Save the trained model for later use (e.g., deployment)
joblib.dump(model, 'health_issue_model.pkl')
```
